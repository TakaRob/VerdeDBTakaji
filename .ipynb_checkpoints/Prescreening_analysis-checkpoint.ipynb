```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import pandas as pd
import numpy as np
from datetime import datetime
%matplotlib widget
```

```python
def epoch_to_timestamp(ts):
    '''Convert a unix epoch to a YYYY-MM-DD hh:mm:ss timestamp'''
    return datetime.fromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
```

```python
# load the database file.  
# I have removed unwanted data, but it is otherwise untouched from the original dataverse download.
df = pd.read_csv("crf3d_jv_test_datas_020525.csv")

# Filter out old data.
df = df[df['crf3d_base_time_sec']>=1738008138.16]

# Select just the columns I need for the analaysis.
df2 = df[['crf3d_sample_id', 'crf3d_cell_number',
         'crf3d_pce_pct', 'crf3d_voc_v', 'crf3d_ff_pct', 'crf3d_jsc_macm2', 'crf3d_scan_type', 'crf3d_test_status']].copy()

# Reconstruct the timestamp from the unix epoch.
tsl = df['crf3d_base_time_sec']+60.0*df['crf3d_elapsed_time_sec']
timestamp_series = tsl.apply(epoch_to_timestamp)

df2.insert(0,"created_on", timestamp_series)
df2.insert(1,"time_sec", tsl-min(df['crf3d_base_time_sec']))

df2
```

```python
# The simplest plot is time 
plt.figure()
plt.plot(df2["time_sec"]/3600, df2["crf3d_pce_pct"], '.')
plt.xlabel("Elapsed time (hours)")
plt.ylabel("PCE (percent)")
plt.title("All prescreening measurements through 020525")
```

```python
# Filter by three criteria
def filter_dataframe(my_df,  time_interval, sample_id, scan_direction='R', mode='extract'):
    '''Filter a dataframe from an epoch time interval, a sample id, and a scan direction.'''
    if mode == 'extract':
        my_df_2 = my_df[(my_df['time_sec'] > time_interval[0]) & (my_df['time_sec'] < time_interval[1])]
        my_df_2 = my_df_2[my_df_2['crf3d_sample_id'] == sample_id]
        my_df_2 = my_df_2[my_df_2['crf3d_scan_type'] == scan_direction]
        print(f"filter_dataframe: returning {len(my_df_2)} measurments.")
        return my_df_2
    elif mode == 'remove':
        my_df_2 = my_df[(my_df['time_sec'] < time_interval[0]) | (my_df['time_sec'] > time_interval[1])]
        # my_df_2 = my_df_2[my_df_2['crf3d_sample_id'] == sample_id]       # mostly I want to exclude by time only...
        # my_df_2 = my_df_2[my_df_2['crf3d_scan_type'] == scan_direction]  # so comment these filters out for now.
        print(f"filter_dataframe: returning {len(my_df_2)} measurments.")
        return my_df_2
    else:
        print(f"filter_dataframe: {mode} is an invalide mode. returning 0 measurments.")
        return pd.DataFrame() # empty dataframe
```

```python
# error boxes function
def make_error_boxes(ax, xdata, xerror, yerror_25, yerror_75, facecolor='r', edgecolor='none', alpha=0.5):
    '''Create boxes with the same x-widths and absolute upper and lower edges.'''
    errorboxes = []
    xe = xerror
    for x, ye25, ye75 in zip(xdata, yerror_25, yerror_75):
        # print(f' x={x}, ye25={ye25}, ye75={ye75} ')
        rect = patches.Rectangle((x - xe, ye25), 2.0*xe, ye75 - ye25, facecolor=facecolor, edgecolor=edgecolor, alpha=alpha)
        errorboxes.append(rect)
    
    for box in errorboxes:
        ax.add_patch(box)
```

```python
# Exclude the bad measurements in the range 13 to 16 hours. 
# The first set on Saturday had the lamp set at 0.72 instead of 0.82 by accident.
# Some of them were marked as inactive, and that might be a better way to filter them in the future.
# Note: the 3rd and 4th parameters are ignored for mode = 'remove'.

def my_timeseriesplot(df2, time_ranges, Sample1, column1, title1, ylabel1, ylim1, heating = 'no'):
    df_BCP_C1_F = filter_dataframe(df2, [3600*0.0, time_ranges[-1][-1]], Sample1, scan_direction='F',mode='extract')
    df_BCP_C1_R = filter_dataframe(df2, [3600*0.0, time_ranges[-1][-1]], Sample1, scan_direction='R',mode='extract')
    
    # extract the data for the averages
    df_BCP_F_list = []
    df_BCP_R_list = []
    
    for timerange in time_ranges:
        df_BCP_F_list.append( filter_dataframe(df2, timerange, Sample1, scan_direction='F',mode='extract') )
        df_BCP_R_list.append( filter_dataframe(df2, timerange, Sample1, scan_direction='R',mode='extract') )
    
    
    avg1 = lambda column, series: np.average(series[column])
    med1 = lambda column, series: np.median(series[column])
    time_F = [avg1('time_sec',my_df)/3600 for my_df in df_BCP_F_list]
    pce_F =  [med1(column1,my_df) for my_df in df_BCP_F_list]
    time_R = [avg1('time_sec',my_df)/3600 for my_df in df_BCP_R_list]
    pce_R =  [med1(column1,my_df) for my_df in df_BCP_R_list]
    
    # extract the data for the box plot
    pct25 = lambda column, series: np.percentile(series[column], 25)
    pct75 = lambda column, series: np.percentile(series[column], 75)
    pct_25_F =  [pct25(column1,my_df) for my_df in df_BCP_F_list]
    pct_75_F =  [pct75(column1,my_df) for my_df in df_BCP_F_list]
     
    pct_25_R =  [pct25(column1,my_df) for my_df in df_BCP_R_list]
    pct_75_R =  [pct75(column1,my_df) for my_df in df_BCP_R_list]
    
    # Figure and axes
    fig, ax = plt.subplots(1)
    params = {'legend.fontsize': 'x-large',
              'figure.figsize': [15, 9],
             'axes.labelsize': 'x-large',
             'axes.titlesize':'x-large',
             'xtick.labelsize':'x-large',
             'ytick.labelsize':'x-large'}
    plt.rcParams.update(params)
    
    # Plot the 25-75 error boxes.
    x_error = 0.8
    make_error_boxes(ax, time_R, x_error, pct_25_R, pct_75_R, edgecolor='b', facecolor='none')
    make_error_boxes(ax, time_F, x_error, pct_25_F, pct_75_F, edgecolor='r', facecolor='none')
    
    # Plot the averages as dashed lines and the individual data points with alpha=0.5. 
    plt.plot(time_R, pce_R, 'b--')
    plt.plot(df_BCP_C1_R["time_sec"]/3600, df_BCP_C1_R[column1], 'bo', alpha=0.5, label='Reverse')
    #
    plt.plot(time_F, pce_F, 'r--')
    plt.plot(df_BCP_C1_F["time_sec"]/3600, df_BCP_C1_F[column1], 'ro', alpha=0.5, label='Forward')
    
    # Mark the positions of the heating steps
    if heating == 'yes':
        y = [0, 100]
        ax.fill_betweenx(y, 19.25, 21.75, color='y', alpha=0.25, label = 'dark, 65Â°C, 2.5 hr.')
        ax.fill_betweenx(y, 40.75, 44.25, color='orange', alpha=0.25, label = 'dark, 85Â°C, 3.5 hr.')
    
    # finshing touches
    plt.xlabel("Elapsed time (hours)")
    plt.ylabel(ylabel1)
    plt.title(title1)
    plt.legend()
    plt.xlim(-2,time_ranges[-1][-1]/3600)
    plt.ylim(ylim1[0],ylim1[1])
    plt.grid('on')

    # Close for later plotting
    plt.close(fig)
    
    return fig
```

```python
# parameters
Sample1 = '01272025_FAPI_SA_Stability comparison_MeO-4PACz_C4'
column1 = 'crf3d_pce_pct'  #  'crf3d_ff_pct' # 
ylabel1 = "PCE (percent)"  # "FF (percent)"
ylim1 = [0,20]
# 
time_ranges = [[3600*0.0, 3600*2.0], [3600*70.0, 3600*75.0] ]
time_ranges = [[3600*0.0, 3600*2.0], [3600*70.0, 3600*73.0], [3600*96.0, 3600*98.0], [3600*163, 3600*168] ]
title1 = f"{Sample1}   "
heating = 'no'
my_timeseriesplot(df2, time_ranges, Sample1, column1, title1, ylabel1, ylim1, heating)
```

```python
def my_boxplot(df2, interval, samples, column, scan_direction, title, ylabel):
    '''Create a customized box plot comparing multiple samples.'''
    # Extract data from the dataframe for a specified time range, sample id, and scan direction.
    df3_1 = filter_dataframe(df2, interval, samples[0], scan_direction=scan_direction)
    data1 = list(df3_1[column])
    
    df3_2 = filter_dataframe(df2, interval, samples[1], scan_direction=scan_direction)
    data2 = list(df3_2[column])
    
    df3_3 = filter_dataframe(df2, interval, samples[2], scan_direction=scan_direction)
    data3 = list(df3_3[column])
    
    df3_4 = filter_dataframe(df2, interval, samples[3], scan_direction=scan_direction)
    data4 = list(df3_4[column])

    # Create the legend strings.
    str1 = lambda name,series: f'{name}: average={round(np.average(series),1)}%, median={round(np.median(series),1)}%, data={list(series)}'
    legends = [str1('C1', df3_1[column]),str1('C2', df3_2[column]),str1('C3', df3_3[column]),str1('C4', df3_4[column])]

    # Figure and axes.
    fig, ax = plt.subplots(1)
    params = {'legend.fontsize': 'medium',
          'figure.figsize': [12, 7],
         'axes.labelsize': 'x-large',
         'axes.titlesize':'x-large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'x-large'}
    plt.rcParams.update(params)
    
    # Creating plot
    bplot = ax.boxplot([data1, data2, data3, data4], tick_labels=[sample1, sample2, sample3, sample4], 
                patch_artist=True)

    # fill with colors
    for patch in bplot['boxes']:
        patch.set_facecolor('lightblue')
    
    # Finishing touches
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(legends)
    plt.ylim(5,20)
    plt.grid('on', alpha=0.25)

    # Close for later plotting
    plt.close(fig)
    
    return fig
```
