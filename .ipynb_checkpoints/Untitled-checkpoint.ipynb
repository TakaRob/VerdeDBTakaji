```python
# Basic imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date, datetime
import os
import json

from db_form import parameters_form 

# Configure matplotlib for inline or widget display
%matplotlib inline 
# OR for interactive plots:
%matplotlib widget 

# Import your custom functions
# Ensure functions.py and db_dataverse.py are in the Python path or same directory
try:
    from functions import (
        epoch_to_timestamp, timestamp_to_epoch, dataframe_clumping,
        read_json_file, get_parameter_details, plot_time_series_from_form,
        my_boxplot_comparison1, # Renamed from plot_box_comparison for consistency if you change it
        get_sample_data_by_criteria_nb, get_jv_data_by_criteria_nb # New functions
    )
    from db_dataverse import SampleDataverseTable, JVScanDataverseTable
    from db_form import parameters_form # If you keep using this for form-based input
except ImportError as e:
    print(f"Error importing custom modules: {e}")
    print("Please ensure functions.py, db_dataverse.py, and db_form.py are accessible.")

# --- Global Configurations ---
CRM_URL = "https://perovskitedevelop.api.crm.dynamics.com/"
SAMPLE_TABLE_LOGICAL_NAME = "crf3d_sample_datas"
JV_SCAN_TABLE_LOGICAL_NAME = "crf3d_jv_test_datas"

# Column logical names for Sample Data (as in your original notebook)
SAMPLE_TABLE_LOGICAL_NAMES = {
    "Sample ID": 'crf3d_sampleid', 
    "Operator": 'crf3d_operatorname', 
    "Perovskite Composition": 'crf3d_perovskitecomposition', 
    "HTL Material": 'crf3d_htlmaterial', 
    "ETL Material": 'crf3d_etlmaterial',
    "Top Capping Material": 'crf3d_topcappingpassivationmaterial', 
    "Bottom Capping Material": 'crf3d_bottomcappingpassivationmaterial', 
    "Bulk Passivation Materials": 'new_bulkpassivationmaterial', 
    "Is Encapsulated": 'crf3d_isencapsulated'
}

# Column logical names for JV Test Data (as in your original notebook)
JV_TABLE_LOGICAL_NAMES = {
    'sample_id': 'crf3d_sample_id',
    'elapsed_time': 'crf3d_elapsed_time_sec',
    'base_time': 'crf3d_base_time_sec',
    'test_id': 'crf3d_test_id',
    'i_ph_suns': 'crf3d_iph_suns',
    'voc_v': 'crf3d_voc_v',
    'mpp_v': 'crf3d_mpp_v',
    'jsc_ma': 'crf3d_jsc_macm2',
    'rsh': 'crf3d_rsh',
    'rser': 'crf3d_rser',
    'ff': 'crf3d_ff_pct',
    'pce': 'crf3d_pce_pct',
    'operator': 'crf3d_operator_name',
    'scan_type': 'crf3d_scan_type',
    'lab_location': 'crf3d_location',
    'cell_number': 'crf3d_cell_number',
    'module': 'crf3d_module',
    'masked': 'crf3d_masked',
    'mask_area': 'crf3d_mask_area_cm2',
    'temp_c': 'crf3d_temperature_c',
    'hum_pct': 'crf3d_humidity_pct',
    'four_wire_mode': 'crf3d_four_wire_mode',
    'scan_data_path': 'crf3d_scan_path'
}

JUPYTER_NOTEBOOK_PATH = os.getcwd() # Or a more specific path if needed
REPORTS_BASE_PATH = os.path.join(JUPYTER_NOTEBOOK_PATH, "../Reports") # Adjusted path

print("Setup Complete. CRM URL:", CRM_URL)
print("Jupyter Notebook Path:", JUPYTER_NOTEBOOK_PATH)
print("Reports Base Path:", REPORTS_BASE_PATH)
```

```python
try:
    sample_table = SampleDataverseTable(
        crm_url=CRM_URL,
        table_name=SAMPLE_TABLE_LOGICAL_NAME,
        col_logical_names=SAMPLE_TABLE_LOGICAL_NAMES
    )
    jv_test_table = JVScanDataverseTable(
        crm_url=CRM_URL,
        table_name=JV_SCAN_TABLE_LOGICAL_NAME,
        col_logical_names=JV_TABLE_LOGICAL_NAMES
    )
    print("Dataverse table objects initialized successfully.")
except Exception as e:
    print(f"Error initializing Dataverse table objects: {e}")
    sample_table = None
    jv_test_table = None
```

```python
# --- User Defined Filters ---
# For sample IDs: Use a list of exact IDs. 
# E.g., specific_sample_ids = ["05072025_CubicPV_VRD-5_dev1", "05072025_Module_M4"]
# Set to None or [] to not filter by specific sample IDs (though usually you'll want some).
#specific_sample_ids = ["05072025_CubicPV_VRD-5_dev1"] # Example
specific_sample_ids = [] # Example


# For date range: Use 'YYYY-MM-DD' format. Set to None to omit a date boundary.
date_from = "2025-05-07"  # Example start date
date_to = "2025-05-07"    # Example end date

# For operator name: Set to None if no operator filter is needed.
# Note: This requires adding operator name filtering to get_entries_by_criteria if desired,
# or filtering the DataFrame after retrieval if 'operatorname' is fetched.
# For now, we'll assume 'operatorname' is in the fetched columns and filter later if needed.
# operator_filter = "Seid" 

# Maximum number of JV scan entries to fetch *after* filtering (can be None for all matches)
jv_fetch_limit = 0 
sample_fetch_limit = 0 # For sample metadata

print(f"Retrieving data for Sample IDs: {specific_sample_ids if specific_sample_ids else 'All matching date/operator'}")
print(f"Date range: {date_from if date_from else 'Any'} to {date_to if date_to else 'Any'}")
```

```python
sample_ids_retrieved = []
sample_details_df = pd.DataFrame()
jv_scan_ids_retrieved = []
jv_scans_df = pd.DataFrame()

if sample_table:
    print("\nFetching sample details...")
    # If you want to fetch details for specific_sample_ids:
    if specific_sample_ids:
         _s_ids, sample_details_df, _s_res = get_sample_data_by_criteria_nb(
            sample_table, 
            sample_ids=specific_sample_ids,
            date_from=None, # Usually sample metadata isn't date-filtered unless creation date is key
            date_to=None,
            limit=sample_fetch_limit 
        )
         if _s_ids:
             sample_ids_retrieved.extend(_s_ids)
         print(f"Retrieved {len(sample_details_df)} sample detail entries.")
    else:
        # Fallback or alternative logic if no specific_sample_ids are given but date range is for samples
        print("No specific sample IDs provided for metadata fetching. Date filtering for samples not implemented here.")
        # You might fetch based on date if that's a use case for sample metadata itself.

if jv_test_table:
    print("\nFetching JV scan data...")
    jv_scan_ids_retrieved, jv_scans_df, _jv_res = get_jv_data_by_criteria_nb(
        jv_test_table,
        sample_ids=specific_sample_ids, # Use the same list of sample IDs
        date_from=date_from,
        date_to=date_to,
        limit=jv_fetch_limit
    )
    print(f"Retrieved {len(jv_scans_df)} JV scan entries.")
    if not jv_scans_df.empty:
        print("JV Scans DataFrame Head:")
        print(jv_scans_df.head())
        # Ensure 'abs_epoch_time' is present after fetching (get_jv_data_by_criteria_nb should handle this)
        if 'abs_epoch_time' not in jv_scans_df.columns:
            print("Warning: 'abs_epoch_time' column is still missing in jv_scans_df after fetching.")
        else:
            print("'abs_epoch_time' column is present.")

# Populate sample_id_list for downstream use (e.g., forms)
# This should be the unique list of sample IDs for which JV data was actually found.
if not jv_scans_df.empty and 'sample_id' in jv_scans_df.columns:
    sample_id_list_for_forms = sorted(list(jv_scans_df['sample_id'].unique()))
    print(f"\nUnique Sample IDs with JV data for form selection: {sample_id_list_for_forms}")
else:
    sample_id_list_for_forms = []
    print("\nNo JV data found to populate sample_id_list_for_forms.")
```

```python
jv_test_data_list_clumped = []
if not jv_scans_df.empty and sample_id_list_for_forms:
    # dataframe_clumping expects a list of DataFrames, one per sample
    jv_data_per_sample_for_clumping = [
        jv_scans_df[jv_scans_df["sample_id"] == sid].copy() for sid in sample_id_list_for_forms
    ]

    min_gap_width_clump = 1.0  # hours
    max_clump_width_clump = 20  # hours

    jv_test_data_list_clumped = dataframe_clumping(
        jv_data_per_sample_for_clumping, 
        min_gap_width_clump, 
        max_clump_width_clump
    )
    print(f"Data clumping complete for {len(jv_test_data_list_clumped)} samples.")
    # You might want to inspect the first clumped DataFrame:
    # if jv_test_data_list_clumped and not jv_test_data_list_clumped[0].empty:
    #     print("\nHead of first clumped DataFrame:")
    #     print(jv_test_data_list_clumped[0].head())
else:
    print("Skipping clumping as no JV data or sample IDs are available.")
```

```python
default_operator = "Unknown"
if jv_test_data_list_clumped and not jv_test_data_list_clumped[0].empty and 'operator' in jv_test_data_list_clumped[0].columns:
    # Try to get operator from the first retrieved and clumped sample's data
    # Ensure there's data and the 'operator' column exists and is not all NaN
    first_operator = jv_test_data_list_clumped[0]['operator'].dropna().iloc[0] if not jv_test_data_list_clumped[0]['operator'].dropna().empty else "Unknown"
    default_operator = str(first_operator)

# You can make this an input field or hardcode for now
operator_name_input = input(f"Operator name (default: {default_operator}): ") or default_operator
print(f"Using Operator: {operator_name_input}")

DIR_PATH_LOCAL_REPORTS = os.path.join(REPORTS_BASE_PATH, operator_name_input) 
if not os.path.exists(DIR_PATH_LOCAL_REPORTS):
    os.makedirs(DIR_PATH_LOCAL_REPORTS, exist_ok=True)

TODAY_DATE = date.today()
CURRENT_REPORT_PATH = os.path.join(DIR_PATH_LOCAL_REPORTS, TODAY_DATE.strftime("%m%d%y"))
if not os.path.exists(CURRENT_REPORT_PATH):
     os.makedirs(CURRENT_REPORT_PATH, exist_ok=True)
print(f"Report Path: {CURRENT_REPORT_PATH}")

# Load stability test conditions (if used)
stability_json_filename = f"{operator_name_input}_degradation_tests_{TODAY_DATE}_V2.json" # Or a fixed name
stability_json_filepath = os.path.join(DIR_PATH_LOCAL_REPORTS, stability_json_filename) 
stability_tests_list = None
if os.path.exists(stability_json_filepath):
    stability_tests_list = read_json_file(stability_json_filepath)
    print(f"Loaded degradation tests from: {stability_json_filepath}")
else:
    print(f"INFO: Degradation test JSON file not found at '{stability_json_filepath}'. Stability spans will not be plotted if this file is needed.")
stability_tests_df_global = pd.DataFrame(stability_tests_list) if stability_tests_list else pd.DataFrame()

# Determine overall min_base_time for consistent time_hrs axis
min_base_time_for_all_plots = None
if not jv_scans_df.empty and 'base_time' in jv_scans_df.columns:
    numeric_base_times = pd.to_numeric(jv_scans_df['base_time'], errors='coerce').dropna()
    if not numeric_base_times.empty:
        min_base_time_for_all_plots = numeric_base_times.min()
```

```python
from db_form import parameters_form # Ensure this is imported if not already
questions_file_main_path = os.path.join(JUPYTER_NOTEBOOK_PATH, "main_questions.json")

if sample_id_list_for_forms: # Only display form if there are samples
    form_interactive = parameters_form(
        persistent=True, 
        dir_path_local=DIR_PATH_LOCAL_REPORTS, # Or CURRENT_REPORT_PATH if specific to date
        questions_file_main=questions_file_main_path, 
        checkbox_values=sample_id_list_for_forms # Use the dynamically fetched list
    )
    print(f"Checkbox values for form: {form_interactive.checkbox_values}")
    form_interactive.display_question_canvas() 
else:
    print("No samples available for form-based plotting.")
    form_interactive = None
```

```python
if form_interactive and form_interactive.responses:
    form_questions = [q['question'] for q in form_interactive.questions]
    form_responses_list = list(form_interactive.responses)
    # Handle sample selection from checkboxes
    sample_selection_question = "Sample names? (select up to four)" # Or actual question from JSON
    if sample_selection_question in form_questions:
        sample_indices_selected = form_responses_list[form_questions.index(sample_selection_question)]
        selected_samples_from_form_interactive = [
            sid for idx, sid in enumerate(sample_id_list_for_forms) if sample_indices_selected[idx]
        ]
        form_responses_list[form_questions.index(sample_selection_question)] = selected_samples_from_form_interactive
    else:
        selected_samples_from_form_interactive = []

    form_responses_dict_interactive = dict(zip(form_questions, form_responses_list))
    print(f"Processed interactive form responses: {form_responses_dict_interactive}")
     # Extract parameters for plotting
    plot_type_resp_interactive = form_responses_dict_interactive.get("Plot type?")
    parameter_ylabel_resp_interactive = form_responses_dict_interactive.get("Parameter to plot?")
    direction_response_resp_interactive = form_responses_dict_interactive.get("Scan direction?")

    if not all([plot_type_resp_interactive, parameter_ylabel_resp_interactive, direction_response_resp_interactive, selected_samples_from_form_interactive]):
        print("Form not fully completed or no samples selected. Skipping form-based plotting.")
    else:
        parameter_col_code_interactive, parameter_label_interactive = get_parameter_details(parameter_ylabel_resp_interactive)

        # Filter the jv_test_data_list_clumped for selected_samples_from_form_interactive
        # This requires mapping selected sample IDs back to their DataFrames in jv_test_data_list_clumped
        # Assuming jv_test_data_list_clumped items correspond to sample_id_list_for_forms
        selected_dfs_for_plotting_interactive = []
        for selected_sample_id in selected_samples_from_form_interactive:
            try:
                idx = sample_id_list_for_forms.index(selected_sample_id)
                if idx < len(jv_test_data_list_clumped):
                    selected_dfs_for_plotting_interactive.append(jv_test_data_list_clumped[idx])
            except ValueError:
                print(f"Warning: Sample ID {selected_sample_id} from form not found in clumped data list.")

        if not selected_dfs_for_plotting_interactive:
            print("No valid data found for the samples selected in the form.")
        else:
            if plot_type_resp_interactive == "Time series":
                # ... (call plot_time_series_from_form) ...
                print(f"Plotting Time Series via form for: {parameter_label_interactive}")
                y_limits_ts_interactive = None # Set specific if needed based on param
                if parameter_col_code_interactive == "pce": y_limits_ts_interactive = [0, 20]

                fig_ts_interactive = plot_time_series_from_form(
                    df_list_processed=selected_dfs_for_plotting_interactive, 
                    sample_ids_selected=selected_samples_from_form_interactive,
                    parameter_to_plot=parameter_col_code_interactive, 
                    parameter_ylabel=parameter_label_interactive, 
                    direction_response=direction_response_resp_interactive,
                    stability_tests_df=stability_tests_df_global,
                    min_base_time_overall=min_base_time_for_all_plots,
                    y_limits_param=y_limits_ts_interactive
                )
                plt.show()
            elif plot_type_resp_interactive == "Box plot":
                # ... (call my_boxplot_comparison1) ...
                print(f"Plotting Box Plot via form for: {parameter_label_interactive}")
                initial_period_hours_bp_interactive = 2.0 
                time_interval_for_initial_bp_interactive = []

                # Determine time interval for box plot using the main jv_scans_df
                # but filter it for the samples selected in the form
                df_for_time_ref_bp_interactive = jv_scans_df[jv_scans_df['sample_id'].isin(selected_samples_from_form_interactive)]

                if not df_for_time_ref_bp_interactive.empty and 'abs_epoch_time' in df_for_time_ref_bp_interactive.columns:
                    min_epoch_bp_interactive = df_for_time_ref_bp_interactive['abs_epoch_time'].dropna().min()
                    if pd.notna(min_epoch_bp_interactive):
                        time_interval_for_initial_bp_interactive = [
                            min_epoch_bp_interactive, 
                            min_epoch_bp_interactive + (initial_period_hours_bp_interactive * 3600)
                        ]
                
                if time_interval_for_initial_bp_interactive:
                    short_labels_bp_interactive = [s.split('_')[-1] if '_' in s else s for s in selected_samples_from_form_interactive]
                    plot_title_bp_interactive = (f"Initial {initial_period_hours_bp_interactive}hrs Performance ({parameter_ylabel_resp_interactive})\n"
                                         f"Comparing: {', '.join(short_labels_bp_interactive)} - {direction_response_resp_interactive} Scans")
                    y_limits_bp_map_val_interactive = {"pce": [0, 25], "voc_v": [0.5, 1.3], "jsc_ma": [0, 30], "ff": [0, 90]}
                    y_limits_val_bp_interactive = y_limits_bp_map_val_interactive.get(parameter_col_code_interactive, [None, None])

                    fig_box_interactive = my_boxplot_comparison1(
                        main_df_with_abs_time=jv_scans_df.copy(), # Pass the main df
                        time_interval_epoch=time_interval_for_initial_bp_interactive,
                        list_of_sample_ids=selected_samples_from_form_interactive,
                        parameter_to_plot=parameter_col_code_interactive,
                        scan_direction_to_plot=direction_response_resp_interactive.upper() if direction_response_resp_interactive.lower() != 'both' else 'Both',
                        plot_title=plot_title_bp_interactive,
                        y_axis_label=parameter_label_interactive,
                        y_limits=y_limits_val_bp_interactive,
                        short_sample_labels=short_labels_bp_interactive
                    )
                    plt.show()
                else:
                    print("Could not generate box plot via form: time interval issue.")
else:
    print("Form not used or responses not available.")
```

```python
# --- Manual Time Series Plot Configuration ---
#manual_ts_sample_ids = ['05072025_CubicPV_VRD-5_dev1'] # Example: single sample
manual_ts_sample_ids = sample_id_list_for_forms[:1] # Example: first two retrieved samples for overview

manual_ts_parameter_label = 'PCE (%)' 
manual_ts_scan_direction = 'Reverse' # 'Forward', 'Reverse', or 'Both'
manual_ts_plot_title = f"Manual Time Series: {manual_ts_parameter_label} for {', '.join(manual_ts_sample_ids)} ({manual_ts_scan_direction} Scans)"
manual_ts_y_limits = [None, None] # e.g., [0, 25] for PCE, or [None, None] for auto

if not jv_test_data_list_clumped or not manual_ts_sample_ids:
    print("Skipping manual time series: No clumped data or no sample IDs specified.")
else:
    manual_ts_param_code, manual_ts_param_ylabel = get_parameter_details(manual_ts_parameter_label)

    # Filter df_list_processed for the selected manual_ts_sample_ids
    selected_dfs_for_manual_ts = []
    actual_manual_ts_ids = []
    for s_id in manual_ts_sample_ids:
        try:
            idx = sample_id_list_for_forms.index(s_id) # Check against samples we have data for
            if idx < len(jv_test_data_list_clumped):
                selected_dfs_for_manual_ts.append(jv_test_data_list_clumped[idx])
                actual_manual_ts_ids.append(s_id)
        except ValueError:
            print(f"Warning: Sample ID {s_id} for manual TS plot not found in retrieved JV data.")

    if selected_dfs_for_manual_ts:
        fig_manual_ts = plot_time_series_from_form(
            df_list_processed=selected_dfs_for_manual_ts,
            sample_ids_selected=actual_manual_ts_ids,
            parameter_to_plot=manual_ts_param_code,
            parameter_ylabel=manual_ts_param_ylabel,
            direction_response=manual_ts_scan_direction,
            stability_tests_df=stability_tests_df_global, # Optional
            min_base_time_overall=min_base_time_for_all_plots, # Optional
            y_limits_param=manual_ts_y_limits,
            plot_title_override=manual_ts_plot_title
        )
        plt.show()
    else:
        print("No data available for the specified manual time series plot samples.")
```

```python
manual_bp_sample_ids = ['05072025_CubicPV_VRD-5_dev4', '05072025_CubicPV_VRD-5_dev1']
manual_bp_parameter_label = 'PCE (%)'
manual_bp_scan_direction = 'Both' # 'Forward', 'Reverse', or 'Both'
manual_bp_plot_title = f"Manual Box Plot: {manual_bp_parameter_label} for selected samples ({manual_bp_scan_direction})"
manual_bp_y_limits = [0, 25]
manual_bp_short_labels = ['dev4', 'dev1'] # Should match manual_bp_sample_ids
# manual_bp_box_colors = ['pink', 'lightblue'] # Optional
# Define time interval for the box plot (e.g., first 2 hours of data)
initial_period_hours_manual_bp = 2.0 
time_interval_manual_bp = []

if not jv_scans_df.empty and 'abs_epoch_time' in jv_scans_df.columns:
    # Filter jv_scans_df for the *manually selected* samples for this plot
    df_for_time_ref_manual_bp = jv_scans_df[jv_scans_df['sample_id'].isin(manual_bp_sample_ids)]
    if not df_for_time_ref_manual_bp.empty:
        min_epoch_manual_bp = df_for_time_ref_manual_bp['abs_epoch_time'].dropna().min()
        if pd.notna(min_epoch_manual_bp):
            time_interval_manual_bp = [
                min_epoch_manual_bp, 
                min_epoch_manual_bp + (initial_period_hours_manual_bp * 3600)
            ]

if not jv_scans_df.empty and manual_bp_sample_ids and time_interval_manual_bp:
    manual_bp_param_code, manual_bp_param_ylabel = get_parameter_details(manual_bp_parameter_label)
    
    fig_manual_bp = my_boxplot_comparison1(
        main_df_with_abs_time=jv_scans_df.copy(), # Pass the main df
        time_interval_epoch=time_interval_manual_bp,
        list_of_sample_ids=manual_bp_sample_ids,
        parameter_to_plot=manual_bp_param_code,
        scan_direction_to_plot=manual_bp_scan_direction.upper() if manual_bp_scan_direction.lower() != 'both' else 'Both',
        plot_title=manual_bp_plot_title,
        y_axis_label=manual_bp_param_ylabel,
        y_limits=manual_bp_y_limits,
        short_sample_labels=manual_bp_short_labels
        # box_colors=manual_bp_box_colors # Optional
    )
    plt.show()
elif jv_scans_df.empty:
    print("Skipping manual box plot: jv_scans_df is empty.")
elif not manual_bp_sample_ids:
    print("Skipping manual box plot: No sample IDs specified.")
elif not time_interval_manual_bp:
    print("Skipping manual box plot: Could not determine time interval (check if selected samples have data).")
```
